# Immutable and Mutable Types
## Immutable Types
* string, integer, float, tuple
* can not be changed by reference
## Mutable Types
* list, dictionary, set
* can be changed by reference

```python
a: int = 3
def abc(n: int) -> None:
    b: list[int] = a # copy by reference
    b: int = 5 # value updated
    print(f"The value of b is: {b}")
abc(a) # passing by a value
print(f"The value of a is: {a}")
```

```python
a: list[int] = [1, 2, 3, 4]
def abc(n: list[int]) -> None:
    b: list[int] = [6, 7, 8] # value updated
    a.append(5) # method applied
    print(f"The value of b is: {b}")
abc(a)
print(f"The value of a is: {a}")
```

```python
a: list[int] = [1, 2, 3, 4]
def abc(n: list[int]) -> None:
    b: list[int] = a # copy by reference
    a.append(5) # method applied
    print(f"The value of b is: {b}")
abc(a)
print(f"The value of a is: {a}")
```

## Error creating

```python
class StudentCard():
    def __init__(self, roll_no, name, age) -> None:
        if age < 18 or age > 60:
            raise StudentAgeError("You are not eligible for this program!")
        self.roll_no = roll_no
        self.name = name
        self.age = age
class StudentAgeError(Exception): # custom error class
        pass
student1 = StudentCard(1, 'Ali', 15)
```

```python
with open("./abc.txt", 'r') as file:
    print(file.readlines()) # read only
```

```python
with open("./abc1.txt", 'w') as file:
    file.write("Pakistan") # writes but over-ride all the previous data
```

```python
with open("./abc1.txt") as file:
    print(file.readlines())
```

```python
with open("./abc1.txt", 'r+') as file:
    file.write("123456")
    print(file.readlines()) # write on on each time a file runs
```

```python
with open("./abc1.txt", 'r+') as file:
    print(file.readlines())
    file.write("asdf")
    file.seek(0) # return the cursor to the start
    print(file.readlines())
```

### r it reads only
### w it writes, if fresh file, write, if old file, over-ride
### r+ it reads and write
### a adds content at the end
### x create new file, if exist, give error
### rb read in binary mood
